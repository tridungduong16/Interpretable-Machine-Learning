```python
def load_data(path, i_exp):
    col =  ["t", "yf", "ycf", "mu0", "mu1" ]
    cov = ["x" + str(i) for i in range(1,26)]
    col = col + cov
    features = cov + ["t"]
    D = np.load(path)
    df = pd.DataFrame(columns=col)

    for i in range(1,26):
        df['x' + str(i)]  = D['x'][:,i-1,i_exp-1]

    df['t']  = D['t'][:,i_exp-1:i_exp]
    df['yf'] = D['yf'][:,i_exp-1:i_exp]
    df['ycf'] = D['ycf'][:,i_exp-1:i_exp]
    df['mu0'] = D['mu0'][:,i_exp-1:i_exp]
    df['mu1'] = D['mu1'][:,i_exp-1:i_exp]
    return df
```

```python
PATH_TRAIN = "/home/dtd/Downloads/dragonnet/result/ihdp/tarnet/{}/baseline/0_replication_train.npz"
PATH_TEST = "/home/dtd/Downloads/dragonnet/result/ihdp/tarnet/{}/baseline/0_replication_test.npz"
PATH_DATA = "/home/dtd/Downloads/dragonnet/result/ihdp/tarnet/{}/simulation_outputs.npz"
DATA_TRAIN = "/home/dtd/Documents/interpretable_machine_learning/Source Code/data/ihdp_npci_1-100.train.npz"
DATA_TEST = "/home/dtd/Documents/interpretable_machine_learning/Source Code/data/ihdp_npci_1-100.test.npz"

import pandas as pd
mae_train_seq = []
mae_test_seq = []
for i in range(0, 100):    
    df_train = np.load(PATH_TRAIN.format(i))
    df_test = np.load(PATH_TEST.format(i))
    
    
    effect_train = np.mean(df_train['q_t1'] - df_train['q_t0'])
    effect_test = np.mean(df_test['q_t1'] - df_test['q_t0'])
    
    data = load_data(DATA_TRAIN, i + 1)
    data_test = load_data(DATA_TEST, i + 1)
    
    true_train = np.mean(data['mu1'] - data['mu0'])
    true_test = np.mean(data_test['mu1'] - data_test['mu0'])
    
    mae_train = abs(true_train - effect_train)
    mae_test = abs(true_test - effect_test)
    mae_train_seq.append(mae_train)
    mae_test_seq.append(mae_test)
    print(mae_train, mae_test)
```
